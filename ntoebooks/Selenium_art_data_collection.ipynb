```python
import os
import time
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
from PIL import Image
from io import BytesIO
```

```python
# List of art forms
art_labels = ['Art Deco',
'Art Nouveau',
'Avant-Garde',
'Bauhaus',
'Cubism',
'Dadaism',
'Futurism',
'Gothic Art',
'Hyperrealism',
'Impressionism',
'Minimalism',
'Neoclassicism Painting',
'Surrealism',
'Pop Art',
'Rococo'
]
```

```python
# Set up Chrome options
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
chrome_options.add_argument("--disable-gpu")

# Set up the Chrome driver
webdriver_service = Service()
driver = webdriver.Chrome(service=webdriver_service, options=chrome_options)
min_width = 100
min_height = 100


# Base directory
base_dir = '/content/data'

# Function to download images
def download_images(query, output_directory):
  search_url = f"https://www.google.com/search?q={query}&tbm=isch"
  driver.get(search_url)

  time.sleep(2)
  for i in range(3):  # Adjust the range for more scrolling

    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    # print('yo')
    time.sleep(2)



  time.sleep(2)
  soup = BeautifulSoup(driver.page_source, 'html.parser')
  img_tags = soup.find_all('img')
  print(query,len(img_tags))

  count = 0

  for i, img in enumerate(img_tags):
      img_url = img.get('src')
      if img_url and img_url.startswith('http'):
          img_data = requests.get(img_url).content
          # Extract the original filename from the URL
          original_filename = os.path.basename(img_url)
          try:
            img = Image.open(BytesIO(img_data))
            if img.width >= min_width and img.height >= min_height:
              with open(f'{output_directory}/{original_filename}.jpg', 'wb') as handler:
                  handler.write(img_data)
              count += 1
          except Exception as e:
            print(e)


  print('Total Image found',count)



for art_form in art_labels:
    art_form_dir = os.path.join(base_dir, art_form)
    if not os.path.exists(art_form_dir):
        print(art_form_dir,'created')
        os.makedirs(art_form_dir)
    download_images(art_form, art_form_dir)
    print('completed', art_form)

# Close the driver
driver.quit()
```
