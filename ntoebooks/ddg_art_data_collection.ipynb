```python
art_labels = ['Art Deco',
'Art Nouveau',
'Avant-Garde',
'Bauhaus', # without art
'Cubism', # without art
'Dadaism', # without art
'Futurism', # without art 300
'Gothic Art',
'Hyperrealism', # without art
'Impressionism', # without art
'Minimalism', # without art
'Neoclassicism', # painting300
'Surrealism',
'Pop Art',
'Rococo' ## without art
]
data_path = "data"
```

```python
if not os.path.exists(data_path):
  os.mkdir(data_path)

for art_type in art_labels:
  print('Now proceeding -> ',art_type)
  search_phrase = art_type

  dest = f'{data_path}/{art_type}'
  if not os.path.exists(dest):
    os.mkdir(dest)




  try:
    art_image_urls = search_images_ddg(search_phrase, max_images=300)
    download_images(dest, urls = art_image_urls)
    print('Successful',search_phrase, len(art_image_urls))

  except:
    print("this could not be done-->",art_type)
```
